```python
# read nc output in folder

import os
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cm

# define nc file
nc_file = 'aeolis_horizontalgradient1.nc'
# nc_file = 'aeolis_verticallayering2.nc'

# read nc_file and terurn dimensions

def read_nc(nc_file):
    # read nc file
    data = nc.Dataset(nc_file, 'r')
    # get dimensions
    x = data['x'].shape
    y = data['y'].shape
    z = data['zb'].shape
    Ct = data['Ct']

    #close nc file
    
    
    return x, y, z, Ct

# print dimensions
x, y, z, Ct = read_nc(nc_file)
print('x:', x, 'y:', y, 'z:', z)

# plot the last timestep of Ct
def plot_Ct(Ct, x, y):

    # plot Ct in 3 rows
    fig, axs = plt.subplots(3, 1, figsize=(10, 10))

    # plot Ct in 3 rows
    for i in range(3):
        axs[i].plot(Ct[-1,i,:,:])
        axs[i].set_title('Ct')
    plt.show()

   

plot_Ct(Ct, x, y)









```
